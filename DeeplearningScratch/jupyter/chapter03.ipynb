```python
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1) # 책에는 [] 빠져있다고 오탈자에 등록되었으나 이게 맞음..
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y축의 범위 지정
plt.show()
```

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
import numpy as np
import matplotlib.pylab as plt 
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x) # 시그모이드 함수로 변경 
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y축 범위 지정
plt.show()
"""
시그모이드란 S자 모양이라는 뜻이다.
"""
```

```python
# ReLU(렐루) 함수
import numpy as np
import matplotlib.pylab as plt 

def ReLU(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = ReLU(x) # 시그모이드 함수로 변경 
plt.plot(x, y)
plt.ylim(-0.1, 5) # y축 범위 지정
plt.show()
# 3장에서만 시그모이드를 사용, 후반부는 주로 렐루를 사용함
```
