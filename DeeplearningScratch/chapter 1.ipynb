```python
import matplotlib.pyplot as plt
x = np.arange(0,6,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격생성
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.xlabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
